```python
def ask_resume_question(vectorstore, question):

    retriever = vectorstore.as_retriever(search_kwargs={"k": 4})
    retrieved_docs = retriever.invoke(question)

    context = "\n\n".join([doc.page_content for doc in retrieved_docs])

    llm = ChatGoogleGenerativeAI(
        model="gemini-2.5-flash",
        temperature=0.2
    )

    prompt = f"""
    You are a professional resume assistant.

    Answer the question strictly using ONLY the information provided in the resume context.
    If the answer is not found in the resume, say:
    "This information is not mentioned in the resume."

    Resume Context:
    {context}

    Question:
    {question}
    """

    response = llm.invoke(prompt)
    return response.content
```

```python
answer = ask_resume_question(vectorstore, 
                              "location?")

print(answer)
```
